```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset from seaborn
iris = sns.load_dataset("iris")
```

```python
# Shape of the dataset
print("Shape of dataset:", iris.shape)

# Column names
print("\nColumn names:", iris.columns.tolist())

# First 5 rows
print("\nFirst 5 rows:")
print(iris.head())

# Info
print("\nDataset Info:")
print(iris.info())

# Summary statistics
print("\nSummary statistics:")
print(iris.describe())
```

```python
# Pairplot with hue to show species
sns.pairplot(iris, hue='species')
plt.suptitle("Scatter Plot Matrix", y=1.02)
plt.show()
```

```python
# Histograms for all numeric features
iris.hist(figsize=(10, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Box plots for each feature grouped by species
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'Box plot of {column}')
plt.tight_layout()
plt.show()
```
